## **Objective** :

### Product , Price, and Anomaly

```sql
SELECT OrderDate, ProductName, Products.UnitPrice, Quantity, Discount,(1-Discount)*Products.UnitPrice AS UnitPriceDiscounted, 
        (1-Discount)*Products.UnitPrice * Quantity AS TotalPrice
FROM Orders

JOIN [Order Details]
ON Orders.OrderID = [Order Details].OrderID 
JOIN Products
ON [Order Details].ProductID = Products.ProductID
WHERE YEAR(OrderDate) = 1997
```

## **Objective** :

### Customer and Market Size

```sql
SELECT OrderDate, ProductName,(1-Discount)*Products.UnitPrice * Quantity AS TotalPrice,ShipName, ShipCountry,
CASE
    WHEN ShipCountry IN ('Argentina', 'Brazil', 'Canada','Mexico', 'USA','Venuzuela') Then 0
    ELSE 1
END AS IsEurophian
FROM Orders
JOIN [Order Details]
ON Orders.OrderID = [Order Details].OrderID 
JOIN Products
ON [Order Details].ProductID = Products.ProductID
WHERE YEAR(OrderDate) = 1997
```

```sql

WITH cte AS (
    SELECT MONTH(Orders.OrderDate) AS Month,SUM([Order Details].UnitPrice * [Order Details].Quantity ) AS TotalPrice, Products.ProductName,
    ROW_NUMBER() OVER(PARTITION BY MONTH(Orders.OrderDate) ORDER BY SUM([Order Details].UnitPrice * [Order Details].Quantity )DESC) AS Rank
    FROM [Order Details]
    JOIN Orders
        ON [Order Details].OrderID = Orders.OrderID
    JOIN Products
        ON [Order Details].ProductID = Products.ProductID
    WHERE YEAR(Orders.OrderDate) = 1997 
    GROUP BY Products.ProductName, MONTH(Orders.OrderDate)
    
) 
SELECT * 
FROM cte
WHERE Rank <= 5
```
