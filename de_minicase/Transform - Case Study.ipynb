## **Objective** :

### Product , Price, and Anomaly

```sql
SELECT OrderDate, ProductName, Products.UnitPrice, Quantity, (1-Discount)*Products.UnitPrice AS UnitPriceDiscounted, 
        (1-Discount)*Products.UnitPrice * Quantity AS TotalPrice
FROM Orders
JOIN [Order Details]
ON Orders.OrderID = [Order Details].OrderID 
JOIN Products
ON [Order Details].ProductID = Products.ProductID
```

```sql
SELECT OrderDate, ProductName, Products.UnitPrice, Quantity, (1-Discount)*Products.UnitPrice AS UnitPriceDiscounted, 
        (1-Discount)*Products.UnitPrice * Quantity AS TotalPrice,ShipName, ShipCountry,
CASE
    WHEN ShipCountry IN ('Argentina', 'Brazil', 'Canada','Mexico', 'USA','Venuzuela') Then 0
    ELSE 1
END AS IsEurophian
FROM Orders
JOIN [Order Details]
ON Orders.OrderID = [Order Details].OrderID 
JOIN Products
ON [Order Details].ProductID = Products.ProductID
```
