```python
jams_agg = bandung_df[2].groupby(['date','street']).agg({'level':'median','median_length':'median','median_delay':'median','median_speed_kmh':'median','total_records':'median'}).reset_index()
jams_agg=jams_agg.rename(columns = {'total_records':'median_record'})
jams_agg['id'] = jams_agg.date + '_' + jams_agg.street
jams_agg['level'] = jams_agg.level.astype('int')
```

```python
sns.heatmap(jams_agg.corr(),annot=True)
```

```python
count_street = jams_agg.groupby('street').agg({'id':'count'}).reset_index()
### Ambil yang total recordnya minimal 1 bulan
list_street = count_street[count_street.id > 30]['street']
len(list_street)
```

```python
filtered_jams_agg = jams_agg[jams_agg.street.isin(list_street)]
sns.heatmap(filtered_jams_agg.corr(),annot=True)
```
